```python
import numpy as np
import names
import pandas as pd
import random
```

```python
#Creating a random dataset:

size = 1000

Cities = ["Xanthi", "Kommotini", "Kavala", "Alexandroupoli", "Serres", "Drama"]
Name = []
Age = []
City = []
NumberOfKids = []
Paycheck = []
for i in range(0,size):
    Name.append(names.get_full_name())
    Age.append(random.randint(24,65))
    City.append(random.choice(Cities))
    NumberOfKids.append(int(abs(random.normalvariate(1.2,1))))
    Paycheck.append(550+500*random.random())

dataset = {
    "Name":Name,
    "Age":Age,
    "City":City,
    "NumberOfKids":NumberOfKids,
    "Paycheck":Paycheck
}

df = pd.DataFrame(dataset, columns=["Name","Age","City","NumberOfKids","Paycheck"])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
fig, ax = plt.subplots(figsize=(15,4))

sns.countplot(df["City"])

plt.xticks(rotation=-45)
ax.set_ylabel('People Count', fontsize = 16)
ax.set_xlabel('Year Built', fontsize = 16)
plt.tick_params(axis='x', which='major', labelsize=13)
plt.tick_params(axis='y', which='major', labelsize=12)
ax.yaxis.tick_right()
```

```python
fig, ax = plt.subplots(figsize=(12,12),)

df["NumberOfKids"].value_counts().plot(kind='pie')

ax.set_ylabel('Number of Kids', fontsize = 14)
```

```python
fig, ax = plt.subplots(figsize=(15, 5))

sns.distplot(df["Paycheck"])

ax.set_title("Paycheck Distribution", fontsize = 16)
ax.set_xlabel("Paycheck Value",fontsize = 12)
plt.tick_params(axis='x', which='major')
ax.yaxis.tick_left()
```

```python
df[(df["Paycheck"]==df["Paycheck"].min()) + (df["Paycheck"]==df["Paycheck"].max())]
```

```python
fig, ax = plt.subplots(figsize=(15, 6))

sns.scatterplot(df['Age'], df['Paycheck'])

ax.set_title('Paycheck per Age', fontsize = 15, loc='center')
ax.set_xlabel('Age', fontsize = 13)
ax.set_ylabel('Paycheck', fontsize = 13)
plt.tick_params(axis='y', which='major', labelsize=12)
plt.tick_params(axis='x', which='major', labelsize=12)
ax.yaxis.tick_left() # where the y axis marks will be
```
