```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../Datasets/Titanic/train.csv")      #Load the csv file into a pandas dataframe
df.head()                                              
```

```python
df.info()
df.describe(include="all")
```

```python
df = df.drop("Cabin",axis=1)
df.describe(include="all")
```

```python
AgeMedian = df["Age"].median()
df["Age"] = df["Age"].apply(lambda x: x if x>0 else AgeMedian)
df.describe(include="all")
```

```python
df[df["Embarked"].isna()]
```

```python
df["Embarked"].value_counts()
```

```python
df["Embarked"][df["Embarked"].isna()]="S"
df.loc[[61,829]]
```

```python
df.describe()
```

```python
df.head()
```

```python
df.groupby('Embarked')['Embarked'].count()
```

```python
df["Embarked"]=df["Embarked"].map({'C':-1,'S':0,'Q':1,-1:-1,0:0,1:1})
df["Sex"]=df["Sex"].map({'male':-1,'female':1,-1:-1,1:1})
df.head()
```

```python
df.reset_index(drop=True,inplace=True)
df = df.drop(["Name","Ticket","PassengerId"],axis=1)
df.head()
```

```python
df["Family"]=df["SibSp"]+df["Parch"]
df.drop(["SibSp","Parch"],axis=1,inplace=True)
df.head()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
plt.subplots_adjust(wspace=0.2,hspace=0.2)
fig.set_figheight(25)
fig.set_figwidth(25)

sns.distplot(df["Age"],ax=axes[0,0])
sns.distplot(df["Fare"],ax=axes[0,1])
sns.distplot(df["Embarked"],ax=axes[1,0])
sns.distplot(df["Family"],ax=axes[1,1])

axes[0,0].set_xlabel(xlabel="Age",fontsize = 20)
axes[0,1].set_xlabel(xlabel="Fare",fontsize = 20)
axes[1,0].set_xlabel(xlabel="Embarked",fontsize = 20)
axes[1,1].set_xlabel(xlabel="Family",fontsize = 20)
axes[0,0].tick_params(axis='y', which='major', labelsize=15)
axes[0,0].tick_params(axis='x', which='major', labelsize=15)
axes[0,1].tick_params(axis='y', which='major', labelsize=15)
axes[0,1].tick_params(axis='x', which='major', labelsize=15)
axes[1,0].tick_params(axis='y', which='major', labelsize=15)
axes[1,0].tick_params(axis='x', which='major', labelsize=15)
axes[1,1].tick_params(axis='y', which='major', labelsize=15)
axes[1,1].tick_params(axis='x', which='major', labelsize=15)
```
